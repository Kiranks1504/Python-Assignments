```python
# 1.Write a Python Program to Find LCM?
#lowest common multiple for 2 numbers only. For multiple numbers.. WIP.

def prod(*args):
    prod = 1
    for i in range(0, len(args)):
        prod = prod * args[i]
    return prod
def lcm(num1, num2):
    num1_divisor = [i for i in range(1, (num1+1)) if num1 % i ==0]
    num2_divisor = [i for i in range(1, (num2+1)) if num2 % i ==0]
    #print(num1_divisor)
    #print(num2_divisor)
    num1_divisor = set(num1_divisor)
    num2_divisor = set(num2_divisor)
    common_divisors = num1_divisor & num2_divisor
    gcd = max(common_divisors)
    product = prod(num1, num2)
    lcm = int(product/gcd)
    #print("common_divisors:", common_divisors)
    #print("lcm:", lcm)
    return lcm


try:
    num1 = int(input("Enter the first number: "))
    num2 = int(input("Enter the second number: "))
    result = lcm(num1, num2)
    print(f"LCM of {num1} and {num2} : {result}")
    
except:
    print("Encountered some error. Please enter valid positive integers as ur input next time,")
```

```python
# 3.Write a Python Program to Convert Decimal to Binary, Octal and Hexadecimal?

def dec_to_bin(num):
    i = 0
    binary_list= []
    binary = []
    while num > 0:
        rem = num % 2
        num = int(num/2)
        binary_list.append(rem)
        i = i + 1    
    binary = binary_list[::-1]
    binary_string = []
    for ech in binary:
        ech = str(ech)
        binary_string.append(ech)
    binary_string = "".join(binary_string)
    return binary_string

try:
    num = int(input("Enter the number: "))
    result = dec_to_bin(num)
    print(f"Binary of {num}: {result}")
    
except:
    print("Encountered some error. Please enter valid integers as your input next time,")  
```

```python
# 4.Write a Python Program To Find ASCII value of a character?

def ascii_code(char):
    ascii_code = ord(char)
    return ascii_code

try:
    char = input("Enter a character: ")
    result = ascii_code(char)
    print(f"ascii code of {char}: {result}")
    
except:
    print("Encountered some error. Please enter a single character as your input next time.")  
```

```python
# 5.Write a Python Program to Make a Simple Calculator with 4 basic mathematical operations?

# num1, operator, num2 = input("Enter the first number, operator and second number seperated by space: ").split(" ")
# print(operator)

def math_oper(num1, operator, num2):    
    if operator == "+":
        res = num1 + num2
    elif operator == "-":
        res = num1 - num2
    elif operator == "/":
        res = num1/num2
    elif operator == "*":
        res = num1 * num2
    return res
try:
    num1, operator, num2 = input("Enter the first number, operator and second number seperated by space: ").split(" ")
    num1 = int(num1)
    num2 = int(num2)
    result = math_oper(num1, operator, num2)
    print(f"{num1} {operator} {num2} = {result}")
    
except:
    print("Encountered some error. Please enter valid integers as your input next time.") 
```
