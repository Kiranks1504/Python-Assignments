```python
# 1.Write a Python program to find sum of elements in list?
def list_sum(lst):
    sum_total = sum(lst)
    return sum_total
length = int(input("Enter the number of elements in your list: "))
lst = [ ]
for i in range(length):
    entry = int(input(f"Enter you entry number {i+1} for your list: "))
    lst.append(entry)
    
result = list_sum(lst)
print("sum is: ", result)
```

```python
# 2.Write a Python program to  Multiply all numbers in the list?
def list_prod(lst):
    if lst == []:
        product = "No elements in list."
    else:
        product = 1
        for ech in lst:
            product = product * ech
    return product


length = int(input("Enter the number of elements in your list: "))
lst = [ ]
for i in range(length):
    entry = int(input(f"Enter you entry number {i+1} for your list: "))
    lst.append(entry)
    
result = list_prod(lst)
print("Product : ", result)
```

```python
# 3.Write a Python program to find smallest number in a list?
def smallest(lst):
    if lst == []:
        smallest_num = "No elements in list."
    else:
        smallest_num = min(lst)
    return smallest_num


length = int(input("Enter the number of elements in your list: "))
lst = [ ]
for i in range(length):
    entry = int(input(f"Enter you entry number {i+1} for your list: "))
    lst.append(entry)
    
result = smallest(lst)
print("smallest element : ", result)
```

```python
# 4.Write a Python program to find largest number in a list?
def largest(lst):
    if lst == []:
        largest_num = "No elements in list."
    else:
        largest_num = max(lst)
    return largest_num


length = int(input("Enter the number of elements in your list: "))
lst = [ ]
for i in range(length):
    entry = int(input(f"Enter you entry number {i+1} for your list: "))
    lst.append(entry)
    
result = largest(lst)
print("Largest element : ", result)
```

```python
# 6.Write a Python program to find N largest elements from a list?
def n_largest(lst,n):
    if lst == []:
        n_largest_num = "No elements in list."
    elif length == 1:
        n_largest_num = f"There is only one element in your list: {lst[0]}"
    else:
    
        lst = sorted(lst)
        lst1 = set(lst)
        lst = list(lst1)
        if n > len(lst):
            n_largest_num = lst[::-1]
        else:
            n_largest_num = lst[-n:]
    return n_largest_num

        
length = int(input("Enter the number of elements in your list: "))
n = int(input("Enter the count of largest elements you want to see: "))
lst = []
for i in range(length):
    entry = int(input(f"Enter you entry number {i+1} for your list: "))
    lst.append(entry)
    
result = n_largest(lst,n)
print(f"{n} largest elements : ", result)
```

```python
# 7.Write a Python program to print even numbers in a list?
def even(num):
    if num % 2 == 0:
        return True
    else:
        return False
    
def even_lst(lst):
    evenlist = [n for n in lst if even(n) == True]
    if evenlist == []:
        evenlist = f"There are no even elements in your list."
    return evenlist


length = int(input("Enter the number of elements in your list: "))
lst = []
for i in range(length):
    entry = int(input(f"Enter you entry number {i+1} for your list: "))
    lst.append(entry)
    
result = even_lst(lst)
print(f"Even numbers in {lst} : ", result)  
```

```python
# 8.Write a Python program to print odd numbers in a List?
def odd(num):
    if num % 2 != 0:
        return True
    else:
        return False
    
def odd_lst(lst):
    oddlist = [n for n in lst if odd(n) == True]
    if oddlist == []:
        oddlist = f"There are no odd elements in your list."
    return oddlist


length = int(input("Enter the number of elements in your list: "))
lst = []
for i in range(length):
    entry = int(input(f"Enter you entry number {i+1} for your list: "))
    lst.append(entry)
    
result = odd_lst(lst)
print(f"Odd numbers in {lst} : ", result)  
```

```python
# 10.Write a Python program to Cloning or Copying a list?
def copy_lst(lst):
    original_list = lst
    copied_list = lst.copy()
    return copied_list, original_list

lst = [1,2,3]
result = copy_lst(lst)
print(f"original_list, copied_list: ",result)
```

```python
# 11.Write a Python program to Count occurrences of an element in a list?
def count_of_elem(lst, elem):
    count_el = lst.count(elem)
    return count_el

lst = [1,2,3,4,5,8,7,6,2,1,3,4,5,6]
elem = int(input("Enter the element you wish to count occurences in your list: "))
result = count_of_elem(lst,elem)
print(f"Number of occurances of {elem} in your list {lst} : {result} times")
```
