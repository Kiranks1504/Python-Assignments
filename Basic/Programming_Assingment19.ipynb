```python
# 1.Question1
# Create a function that takes a string and returns a string in which each character is repeated once.
# Examples
# double_char("String") ➞ "SSttrriinngg"

# double_char("Hello World!") ➞ "HHeelllloo  WWoorrlldd!!"

# double_char("1234!_ ") ➞ "11223344!!__  "

def double_char(char):
    func = lambda ech: ech*2
    doubles = [func(ech) for ech in char]
    doubles = "".join(doubles)
    return doubles

result = double_char("String")
print(result)
result = double_char("Hello World!")
print(result)
result = double_char("1234!_ ")
print(result)
```

```python
# 2.Question2
# Create a function that reverses a boolean value and returns the string "boolean expected" if another variable type is given.
# Examples
# reverse(True) ➞ False

# reverse(False) ➞ True

# reverse(0) ➞ "boolean expected"

# reverse(None) ➞ "boolean expected"

def reverse(inp_boolean):
    if type(inp_boolean) == bool:
        if inp_boolean == True:
            return False
        elif inp_boolean == False:
            return True
    else:
        return "boolean expected"
    
    
result = reverse(True)
print(result)   
result = reverse(False)
print(result)   
result = reverse(0)
print(result)   
result = reverse(None)
print(result)   
```

```python
# 3.Question3
# Create a function that returns the thickness (in meters) of a piece of paper after folding it n number of times. The paper starts off with a thickness of 0.5mm.
# Examples
# num_layers(1) ➞ "0.001m"
# # Paper folded once is 1mm (equal to 0.001m)

# num_layers(4) ➞ "0.008m"
# # Paper folded 4 times is 8mm (equal to 0.008m)

# num_layers(21) ➞ "1048.576m"
# # Paper folded 21 times is 1048576mm (equal to 1048.576m)

def num_layers(n):
    thickness = 0.0005
    layers_func = lambda x: 2**x
    thickness = layers_func(n) * thickness
    return thickness

n = int(input("Enter the number of times paper is folded: "))
result = num_layers(n)
print(f"Paper folded {n} times is {result}m thick.")
```

```python
# 4.Question4
# Create a function that takes a single string as argument and returns an ordered list containing the indices of all capital letters in the string.
# Examples
# index_of_caps("eDaBiT") ➞ [1, 3, 5]

# index_of_caps("eQuINoX") ➞ [1, 3, 4, 6]

# index_of_caps("determine") ➞ []

# index_of_caps("STRIKE") ➞ [0, 1, 2, 3, 4, 5]

# index_of_caps("sUn") ➞ [1]

def index_of_caps(strng):
    strng = strng[1:-1]
    ind_list = [ ]
    length = len(strng)
    for i in range(0,length):
        char = strng[i]
        uchar = char.upper()
        if char == uchar:
            ind_list.append(i)
        else:
            continue
    return ind_list

strng = input("Enter the string: ")
result = index_of_caps(strng)
print(f"Index of upper case letters: {result}")
```

```python
# 5.Question5
# Using list comprehensions, create a function that finds all even numbers from 1 to the given number.
# Examples
# find_even_nums(8) ➞ [2, 4, 6, 8]

# find_even_nums(4) ➞ [2, 4]

# find_even_nums(2) ➞ [2]

def find_even_nums(num):
    even_nums = [x for x in range(1, num+1) if x % 2 ==0]
    return even_nums
    
num = int(input("Enter the number: "))
result = find_even_nums(num)
print(f"Even numbers from 1 to {num}: {result}")
```
