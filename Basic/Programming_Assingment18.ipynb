```python
# 1.Question 1
# Create a function that takes a list of non-negative integers and strings and return a new list without the strings.
# Examples
# filter_list([1, 2, "a", "b"]) ➞ [1, 2]

# filter_list([1, "a", "b", 0, 15]) ➞ [1, 0, 15]

# filter_list([1, 2, "aasf", "1", "123", 123]) ➞ [1, 2, 123]


def filter_list(lst):
    nw_lst = [i for i in lst if type(i) == int]
    return nw_lst

# lst = [1, 2, "a", "b"]
# lst = [1, "a", "b", 0, 15]
lst = [1, 2, "aasf", "1", "123", 123]

result = filter_list(lst)
print(f"Result of filter_list function on {lst}: {result}")
```

```python
# 2.Question 2
# The "Reverser" takes a string as input and returns that string in reverse order, with the opposite case.
# Examples
# reverse("Hello World") ➞ "DLROw OLLEh"

# reverse("ReVeRsE") ➞ "eSrEvEr"

# reverse("Radar") ➞ "RADAr"

def reverse(strng):
    strng_rev = strng[::-1]
    result_strng = ""
    for i in strng_rev:
        if i.islower():
            i = i.upper()
        elif i.isupper():
            i = i.lower()
        result_strng = result_strng + (i)
    return result_strng    


strng = input("Enter your string: ")
result = reverse(strng)
print(f"Result of the function, reverse() on your string, {strng}: {result}")
```

```python
# 4.Question 4
# Write a function that calculates the factorial of a number recursively.
# Examples
# factorial(5) ➞ 120

# factorial(3) ➞ 6

# factorial(1) ➞ 1

# factorial(0) ➞ 1

def rec_factorial(num):
    fact = 1
    if num>0:
        fact = num * rec_factorial(num-1)
    return fact

num = int(input("Enter the number: "))
result = rec_factorial(num)
print(f"Factorial of {num} found using the recursive function, rec_factorial(): {result}")
```

```python
# 5.Question 5
# Write a function that moves all elements of one type to the end of the list.
# Examples
# move_to_end([1, 3, 2, 4, 4, 1], 1) ➞ [3, 2, 4, 4, 1, 1]
# # Move all the 1s to the end of the array.

# move_to_end([7, 8, 9, 1, 2, 3, 4], 9) ➞ [7, 8, 1, 2, 3, 4, 9]

# move_to_end(["a", "a", "a", "b"], "a") ➞ ["b", "a", "a", "a"]

def move_to_end(lst, ele):
    count = lst.count(ele)
    nw_lst = [x for x in lst if x != ele]
    res_lst = nw_lst + [ele for x in range(1, count+1)]
    return (res_lst)

# lst, ele = [1, 3, 2, 4, 4, 1], 1
# lst, ele = [7, 8, 9, 1, 2, 3, 4], 9
lst, ele = ["a", "a", "a", "b"], "a"

result = move_to_end(lst, ele)
print(f"Result: {result}")
```
